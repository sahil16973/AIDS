```python
from copy import deepcopy
def checkElementHelper(x, S):
   for e in S:
       if x==e[1]: return e[0]
   return 0
```

```python
def union(setA, setB):
   X,Y = deepcopy(setA),deepcopy(setB)
   Z = []
   for i in X:
       mb = checkElementHelper(i[1],Y)
       Z.append( [max(mb, i[0]), i[1]] )
       if mb!=0:
           Y.remove([mb,i[1]])
   Z = Z + Y
   return Z
```

```python
def intersection(setA, setB):
   X,Y = deepcopy(setA),deepcopy(setB)
   Z = []
   for i in X:
       mb = checkElementHelper(i[1],Y)
       if min(mb, i[0])!=0:
           Z.append( [min(mb, i[0]), i[1] ])
   return Z
```

```python
def complement(setA):
   Z = deepcopy(setA)
   for i in Z:
       i[0] = 1-i[0]
       if i[0]==0:
           Z.remove(i)
   return Z
```

```python
def fuzzy_sum(setA, setB):
   X,Y = deepcopy(setA),deepcopy(setB)
   Z = []
   for i in X:
       mb = checkElementHelper(i[1],Y)
       if mb!=0:
           m = round( i[0]+mb-(i[0]*mb), 3)
           Z.append( [m, i[1]] )
           Y.remove([mb, i[1]])
       else:
           Z.append(i)
   Z = Z + Y
   return Z
```

```python
#Algebraic Product
def fuzzy_product(setA, setB):
   X,Y = deepcopy(setA),deepcopy(setB)
   Z = []
   for i in X:
       mb = checkElementHelper(i[1],Y)
       if mb!=0:
           m = i[0]*mb
           Z.append( [m, i[1]] )
           Y.remove([mb, i[1]])

   return Z
```

```python
def bounded_sum(setA, setB):
   X,Y = deepcopy(setA),deepcopy(setB)
   Z = []
   for i in X:
       mb = checkElementHelper(i[1],Y)
       if mb!=0:
           m = i[0]+mb
           Z.append( [min(1,m), i[1]] )
           Y.remove([mb, i[1]])
       else:
           Z.append(i)
   Z = Z + Y
   return Z
```

```python
def bounded_diff(setA, setB):
   X,Y = deepcopy(setA),deepcopy(setB)
   Z = []
   for i in X:
       mb = checkElementHelper(i[1],Y)
       if mb!=0:
           m = round( i[0]-mb, 3)
           Z.append([max(0,m), i[1]] )
           Y.remove([mb, i[1]])
       else:
           Z.append(i)
   return Z
```
