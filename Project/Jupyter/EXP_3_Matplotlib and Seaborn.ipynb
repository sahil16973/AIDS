```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\Users\HP\Downloads\datamind-main\Datasets\iris (1).csv')
```

```python
df.head()
```

```python
df1 = df.drop('species', axis=1)
```

```python
df1.head()
```

```python
corr = df1.corr()
sns.heatmap(corr, annot=True)
```

```python

plt.hist(df['sepal_length'],bins=10,edgecolor='black',color='cyan')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length')
plt.show()
```

```python
# Step 5: Scatter Plot using Matplotlib
plt.scatter(df['sepal_length'], df['petal_length'], color='red')
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
sns.pairplot(df,hue='species')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

```python
sns.boxplot(x='species' , y='sepal_length' , data=df,palette='pastel')
plt.title("Boxplot")
plt.show()
```

```python
plt.plot(df['sepal_length'])
plt.title('Line Plot of Sepal Length')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
plt.bar(df['species'],df['sepal_length'])
plt.xlabel('Species')
plt.ylabel('sepal_length')
plt.show()
```

```python
sns.barplot(x='species', y='sepal_length', data=df)
plt.title('Average Sepal Length by Species')
plt.show()
```
